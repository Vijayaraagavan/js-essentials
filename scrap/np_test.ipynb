```python
import time

def inMilli(a):
    return str(round(a*1000, 2)) + "ms"

def start(s):
    # s = 2_00_000_00
    s1 = time.time()
    arr  = np.random.rand(s)
    np.multiply(arr, 2)
    s2 = time.time() - s1

    s3 = time.time()
    arr4 = list(range(s))
    arr4 = [x * 2 for x in arr4]
    s4 = time.time() - s3

    # print(inMilli(s2), "     ", inMilli(s4))
    return s2, s4

# print("nump    |   list")
l = [10, 20, 1000, 10000, 2000000, 20_000_000, 50_000_000]
r = []
for x in l:
    sm = [x, [], []]
    for _ in range(5):
        s1, s2 = start(x)
        sm[1].append(s1)
        sm[2].append(s2)
    r.append(sm)

print(r)


```

```python
import matplotlib.pyplot as plt
rm = len(r) // 3 + 1
fig, axes = plt.subplots(rm, 3, figsize=(10, 6))  # 2 rows, 2 columns
# fig, axes = plt.subplots(rm, 3)  # 2 rows, 2 columns


for i, n in enumerate(r):
    # plt.subplot(i//3 + 1, len(r), i+1)
    x1 = i//3
    x2 = i%3
    axes[x1, x2].plot(n[1], marker='o', linestyle='-', label="numpy", color='r')
    axes[x1, x2].plot(n[2], marker='o', linestyle='-', label="list")
    axes[x1, x2].set_xlabel("iteration")
    axes[x1, x2].set_ylabel("time")
    axes[x1, x2].legend()
    axes[x1, x2].set_title(str(n[0]) + " input")
plt.tight_layout()
```
